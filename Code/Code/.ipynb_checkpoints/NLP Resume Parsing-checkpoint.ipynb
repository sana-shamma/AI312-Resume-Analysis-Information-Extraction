## Step (1): Import Libraries üìö

```python

# It is used to load serialized data.
import pickle 
# It used to split the dataset into training and testing sets.
from sklearn.model_selection import train_test_split
# It is used for natural language processing tasks.
import spacy
# It is used for random operations such as shuffling data.
import random
# It is used for working with PDF files (such as reading).
import fitz  
```

## Step (2): Loading the Data üìÇ

```python

# Load training data from a pickle file, 'train_data.pkl' contains serialized training data for an NLP model.
inital_data = pickle.load(open('data\\train data\\train_data.pkl', 'rb'))

# Print the first item of the training data to inspect its structure and content.
inital_data[0]
```

```python

# Split the data into training and testing sets.
train_set, test_set = train_test_split(inital_data, test_size=0.05, random_state=42)  # 5% of the data as test set.
```

## Step (3): Train the Model üë®‚Äçüíª

```python
# Initialize a blank English language model in spaCy.
nlp = spacy.blank('en')

def train_model(train_set):
    # Check if a Named Entity Recognition (NER) component does not exist in the pipeline.
    if 'ner' not in nlp.pipe_names:
        # Create a new NER pipe for the model.
        ner = nlp.create_pipe('ner')
        # Add the newly created NER pipe to the pipeline, placing it at the end.
        nlp.add_pipe(ner, last=True)
    
    # Iterate over the training data to add entity labels to the NER component.
    for _, annotation in train_set:
        for ent in annotation['entities']:
            # For each entity, add its label to the NER component for training.
            ner.add_label(ent[2])
    
    # Identify other pipes in the pipeline to disable them during the training of the NER.
    other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
    # Disable all other pipes when training the NER to prevent them from affecting the training.
    with nlp.disable_pipes(*other_pipes):
        # Start training the NER component.
        optimizer = nlp.begin_training()
        # Iterate through multiple iterations of training.
        for itn in range(250):
            print("Starting iteration " + str(itn))
            # Shuffle the training data at the beginning of each iteration to improve learning.
            random.shuffle(train_set)
            # Initialize a dictionary to keep track of the losses.
            losses = {}
            
            # Iterate over each example in the training data.
            for text, annotations in train_set:
                try:
                    # Update the model with the texts and annotations.
                    nlp.update(
                        [text],  # batch of texts
                        [annotations],  # batch of annotations
                        drop=0.2,  # dropout rate to prevent model from memorizing the data
                        sgd=optimizer,  # stochastic gradient descent optimizer
                        losses=losses)
                except Exception as e:
                    # If an error occurs during model update, pass and continue.
                    pass
            
            # Print the losses after each iteration to monitor the training progress.
            print(losses)
```

```python
# Trains the model using the provided training data.
train_model(train_set)
```

## Step (4): Evaluate the Model üìä

```python

def evaluate_model(nlp, test_data):
    # Initialize counters for correct predictions, total predictions, and total correct entities.
    correct_preds, total_preds, total_correct = 0, 0, 0

    # Iterate over each example in the test data.
    for text, annotations in test_data:
        # Process the text using the NLP model to obtain a document object.
        doc = nlp(text)

        # Extract entities recognized by the model as a list of tuples (start position, end position, label).
        true_ents = [(ent.start_char, ent.end_char, ent.label_) for ent in doc.ents]

        # Extract the true (gold standard) entities from the annotations as a list of tuples.
        gold_ents = [(ent[0], ent[1], ent[2]) for ent in annotations['entities']]

        # Convert lists of entities to sets for comparison.
        true_set = set(true_ents)
        gold_set = set(gold_ents)

        # Calculate the number of correct predictions by finding the intersection of predicted and true entities.
        correct_preds += len(true_set & gold_set)

        # Count the total number of predictions made by the model and the total number of true entities.
        total_preds += len(true_set)
        total_correct += len(gold_set)
    
    # Calculate precision as the ratio of correct predictions to total predictions.
    precision = correct_preds / total_preds if total_preds > 0 else 0

    # Calculate recall as the ratio of correct predictions to the total actual entities.
    recall = correct_preds / total_correct if total_correct > 0 else 0

    # Calculate F-score, which is the harmonic mean of precision and recall.
    f_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    # Return the calculated metrics as a dictionary.
    return {"precision": precision, "recall": recall, "f_score": f_score}
```

```python

# Evaluate the model: printing the accuracy of the model that was trained on 190 resumes and tested on 10
results = evaluate_model(nlp, train_set)
print('training set')
print(f"Precision: {results['precision']}")
print(f"Recall: {results['recall']}")
print(f"F-Score: {results['f_score']}\n")

results = evaluate_model(nlp, test_set)
print('test set')
print(f"Precision: {results['precision']}")
print(f"Recall: {results['recall']}")
print(f"F-Score: {results['f_score']}")
```

```python
# save the model.
nlp.to_disk('nlp_model')
```
